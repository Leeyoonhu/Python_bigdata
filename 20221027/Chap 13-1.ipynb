```python
import pandas as pd
```

```python
# csv의 헤더가 없으므로 none
df = pd.read_csv('deep-learning/data/sonar3.csv', header=None)
df.head()
# 일반돌 0, 광석 1
```

```python
df[60].value_counts()
```

```python
X=df.iloc[:, :60]
y=df.iloc[:, 60]
```

```python
from tensorflow.keras import models, layers
model = models.Sequential()
model.add(layers.Dense(24, input_dim=60, activation='relu'))
model.add(layers.Dense(10, activation = 'relu'))
model.add(layers.Dense(1, activation = 'sigmoid'))
```

```python
#(60 x w) * (w * 24) + 24 = 1464
#(24 x w) * (w * 10) + 10 = 250
#(10 x w) * (w * 1) + 1 = 11
#total =1464+250+11 = 1725
model.summary()
```

```python
# 손실함수와 optimizer 장착
model.compile(loss='binary_crossentropy', optimizer='adam', metrics='accuracy')
```

```python
history=model.fit(X, y, epochs=200, batch_size=10)
# 과적합으로 인해 정확도 100%가 나옴
# 과적합을 처리하기위해 모델 성능 검증(p169)
```
