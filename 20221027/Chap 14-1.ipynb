```python
from tensorflow.keras import models, layers
from sklearn.model_selection import train_test_split
import pandas as pd
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
#에포크마다 모델저장
df = pd.read_csv('deep-learning/data/wine.csv', header=None)
X = df.iloc[:, :12]
y = df.iloc[:, 12]
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)
```

```python
model = models.Sequential()
model.add(layers.Dense(30, input_dim=12, activation='relu'))
model.add(layers.Dense(12, activation='relu'))
model.add(layers.Dense(8, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# checkpoint 객체 생성
modelpath = 'deep-learning/data/model/all/{epoch:02d}={val_accuracy:.4f}.hdf5'
# 모델을 저장할때마다 쓰는게 모델 체크포인트
checkpointer = ModelCheckpoint(filepath=modelpath, verbose=1)
# 수다쟁이 verbose =1 >> 로그 출력
history = model.fit(X_train, y_train, epochs=50, batch_size=500, validation_split=0.25, verbose=0, callbacks=[checkpointer])
```

```python
score = model.evaluate(X_test, y_test)
print("Test Accuracy : " + score[1])
```
